#### Non Max Suppression

```
# Calculate the intersection over union
def iou(boxes1, boxes2):
    # make sure the boxes are of 2 dimensions
    if len(boxes1.shape) == 1: boxes1 = np.expand_dims(boxes1, axis=0)
    if len(boxes2.shape) == 1: boxes2 = np.expand_dims(boxes2, axis=0)

    # Calulate the intersectons and unions of of all the boxes in boxes1 with boxes 2
    intersection = np.maximum(0, np.minimum(boxes1[:,2], boxes2[:,2]) - np.maximum(boxes1[:,0], boxes2[:,0])) * np.maximum(0, np.minimum(boxes1[:,3], boxes2[:,3]) - np.maximum(boxes1[:,1], boxes2[:,1]))
    union = (boxes1[:,2] - boxes1[:,0]) * (boxes1[:,3] - boxes1[:,1]) + (boxes2[:,2] - boxes2[:,0]) * (boxes2[:,3] - boxes2[:,1]) - intersection

    return intersection/union

# Calculate the non maximum suppression
def nms(boxes, iou_threshold = 0.5):
    boxes_left = np.copy(boxes)
    maxima = []
    while len(boxes_left) > 0:
        maximum_index = np.argmax(boxes_left[:, 0])
        maximum_box = np.copy(boxes_left[maximum_index])
        maxima.append(maximum_box)
        boxes_left = np.delete(boxes_left, maximum_index, axis = 0)

        if len(boxes_left) == 0:break
        similarities = iou(boxes_left[:,1:], maximum_box[1:])
        boxes_left = boxes_left[similarities <= iou_threshold]

    return np.array(maxima)
```

#### Draw Bounding Boxes

Boxes drawn with raw detection coordinates i.e coordinates/boxes that have not gone through any non-max-suppression

```
def draw_boxes_raw(image, boxes, class_names, scores, social_distant = 150, max_boxes = 10, min_score = 0.35, area_threshold = 1000):

  boxes = np.concatenate([np.expand_dims(scores, axis=1), boxes], axis = 1)
  imheight, imwidth, _ = np.shape(image)

  #Keep boxes with class name index 0
  boxes_person= boxes[np.array(class_names) == 0]
  
  boxes_drawn = nms(boxes_person, iou_threshold=0.2)
  boxes_drawn = boxes_drawn[boxes_drawn[:,0] >= min_score]

  image = np.uint8(image)
  for i in range(len(boxes_drawn)):
    
    ymin, xmin, ymax, xmax = boxes_drawn[i,1:]
    left, right, top, bottom = int (xmin*imwidth), int (xmax*imwidth), int (ymin*imheight), int (ymax*imheight)

    if ((right - left) * (bottom - top)) < area_threshold:
        cv2.rectangle(image, (left, top), (right, bottom), (250,0,0)) 
      

  return image
```

####Get the size of the model

```
# Get model size in bytes
def get_model_size(file_path):
    size = os.path.getsize(file_path)
    return size
# Convert the size in required unit ("bytes, kilobytes, megabytes")
def convert_bytes(size, unit = None):
    if unit == 'KB':
        return print("File size " + str (round(size/1024, 3))  + " kilobytes")

    elif unit == 'MB':
        return print("File size " + str (round(size / (1024 * 1024), 3)) + " megabytes")

    else:
        return print("File size " + str (size) + " bytes")
```

####Model Conversion to its TFLite Equivalent

```
def convert_model_to_tflite(model):
    tf_lite_model_name = "tf_lite_model.tflite"

    tf_lite_converter = tf.compat.v1.lite.TFLiteConverter.from_frozen_graph(
    graph_def_file = model, 
    input_arrays=['normalized_input_image_tensor'],
    output_arrays=['TFLite_Detection_PostProcess','TFLite_Detection_PostProcess:1','TFLite_Detection_PostProcess:2','TFLite_Detection_PostProcess:3'],
    input_shapes={'normalized_input_image_tensor': [1, 320, 320, 3]})

    #tf_lite_converter = tf.lite.TFLiteConverter.from_saved_model(saved_model_path)
    tf_lite_converter.allow_custom_ops = True
    tf_lite_converter.target_spec.supported_types = [tf.float16]
    tf_lite_converter.optimizations = [tf.lite.Optimize.DEFAULT]
    tflite_model = tf_lite_converter.convert()

    with open(tf_lite_model_name, "wb") as f:
      f.write(tflite_model)

    return tflite_model
```

```
# Get the converted tflite model
tflite_model = convert_model_to_tflite("ssdlite_mobiledet_cpu_320x320_coco_2020_05_19/tflite_graph.pb")
```

####TFLite Model Prediction 

```
def get_output_tensor(interpreter, index):
  """Returns the output tensor at the given index."""
  output_details = interpreter.get_output_details()[index]
  tensor = np.squeeze(interpreter.get_tensor(output_details['index']))
  return tensor

def predict_lite(tflite_model, interpreter, image, area_threshold = 1000):
    image = cv2.resize(image, (320, 320))
    img = image.astype('float32') / 255.0
    img = np.expand_dims(img, axis = 0)


    interpreter.allocate_tensors()

    input_details = interpreter.get_input_details()

    interpreter.set_tensor(input_details[0]['index'], img)

    interpreter.invoke()
    boxes = get_output_tensor(interpreter, 0)
    classes = get_output_tensor(interpreter, 1)
    scores = get_output_tensor(interpreter, 2)

    # Return image with the bounding boxes drawn
    image_with_boxes = draw_boxes_raw(np.array(image), boxes, classes, scores, area_threshold = area_threshold)
    
    return image_with_boxes
```

### Prediction of Video Data
    

```
# loop over frames from the video file stream
def video_data_tflite(tflite_model, interpreter, path, writer = None):
  video = cv2.VideoCapture(path)
  fps = FPS().start()
  while (True):
    ret, frame = video.read()

    if not ret :
        break
    image_with_boxes = predict_lite(tflite_model, interpreter, frame)
    # update the FPS counter
    fps.update()
 
    # check if the video writer is None
    if writer is None:
        # initialize our video writer
        fourcc = cv2.VideoWriter_fourcc(*"MJPG")
        writer = cv2.VideoWriter("output.avi", fourcc, 30,
            (image_with_boxes.shape[1], image_with_boxes.shape[0]), True)

    # write the output frame to disk
    writer.write(image_with_boxes)
    if cv2.waitKey(1) == ord('q'):
        break

# stop the timer and display FPS information
  fps.stop()
  print("[INFO] elasped time: {:.2f}".format(fps.elapsed()))
  print("[INFO] approx. FPS: {:.2f}".format(fps.fps()))
    

  

  video.release()
  writer.release()
```

```
# initialize the interpreter
interpreter = tf.lite.Interpreter(model_content = tflite_model)
```
